```python
# performing linear algebra
import numpy as np 

# data processing
import pandas as pd

# visualisation
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./augmented_data.csv")
data
```

```python
data['diagnosis'].value_counts()
```

```python
data
```

```python
data = data.drop(['Unnamed: 32', 'id'], axis = 1)
data
```

```python
data.diagnosis = [1 if each == 'M' else 0 for each in data.diagnosis]
data
```

```python
data['diagnosis'].value_counts()
```

```python
y = data.diagnosis.values
```

```python
x_data = data.drop(['diagnosis'], axis = 1)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x=scaler.fit_transform(x_data)
#x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
	x, y, test_size = 0.15, random_state = 42)

x_train = x_train.T
x_test = x_test.T
y_train = y_train.T
y_test = y_test.T

print("x train: ", x_train.shape)
print("x test: ", x_test.shape)
print("y train: ", y_train.shape)
print("y test: ", y_test.shape)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression

# Impute missing values
imputer = SimpleImputer(strategy='mean')  # You can choose 'mean', 'median', 'most_frequent', or 'constant'
x_train_imputed = imputer.fit_transform(x_train.T)
x_test_imputed = imputer.transform(x_test.T)

# Fit the model
logreg = LogisticRegression(random_state=42, max_iter=150)
print("test accuracy: {}".format(
    logreg.fit(x_train_imputed, y_train.T).score(x_test_imputed, y_test.T)))
print("train accuracy: {}".format(
    logreg.fit(x_train_imputed, y_train.T).score(x_train_imputed, y_train.T)))
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

# Assuming x_train_imputed, y_train, x_test_imputed, y_test are already defined

# Initialize the Logistic Regression model
logreg = LogisticRegression(random_state=42, max_iter=150)

# Train the model
logreg.fit(x_train_imputed, y_train.T)

# Evaluate the model
train_accuracy = logreg.score(x_train_imputed, y_train.T)
test_accuracy = logreg.score(x_test_imputed, y_test.T)

print("train accuracy: {}".format(train_accuracy))
print("test accuracy: {}".format(test_accuracy))
```

```python
y_pred = logreg.predict(x_test_imputed)

# Map predictions to 'M' for malignant (1) and 'B' for benign (0)
y_pred_mapped = np.where(y_pred == 1, 'M', 'B')

# Print the predictions
print("Predictions for x_test: {}".format(y_pred_mapped))
```
